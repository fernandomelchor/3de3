```python
import pandas as pd
import numpy as np
import json
```

```python
#df = pd.read_excel('Data/Merino declaracion_intereses.xlsx',index_col=0, header=None)
df = pd.read_excel('Data/02_DI_JAPM.xlsx',index_col=0, header=None)
df.reset_index(inplace=True, drop=True)
```

```python
def all_null(row):
    if len(row.unique())==1:
        if np.isnan(row.unique()[0]):
            return 'drop'
    return 'keep'
```

```python
df['act'] = df.apply(lambda row: all_null(row), axis=1)
df = df[df['act']=='keep']
df.reset_index(inplace=True, drop=True)
#print(df)
```

```python
tables_num = ['I.','II.','III.','IV.','V.', 'VI.']
def sections(x):
    if (type(x)!=float)&(type(x)!=int):
        for i in tables_num:
            #print(x, i)
            if x.split(' ')[0]==i:
                return i
    return np.nan
df['section'] = df[1].apply(lambda x: sections(x))   
```

```python
x='drop'
x_l=[]

for i in df['section'].tolist():
    if type(i)==str:
        x=i
    x_l.append(x)

df['section']=x_l
```

```python
def check_empty(d):
    l_t = []
    if d.shape[0]==0:
        for i in range(d.shape[1]):
            l_t.append(np.nan)
        d.loc[0] = l_t
    return d
```

```python
def put_nulls(df):
    for col in df:
        df[col]=df[col].str.replace(r"^(\[)(.*?)(\])$",'null')
    return df
```
